```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset (asegurarse de tenerlo en el directorio de trabajo)
file_path = "./data/big_startup_secsees_dataset.csv"  # Cambiar a la ruta correcta
startups_data = pd.read_csv(file_path)

# Analizar las principales m√©tricas y valores del dataset
# Resumen estad√≠stico
print("Resumen estad√≠stico del dataset:")
print(startups_data.describe(include='all'))

# Distribuci√≥n de estados (status)
plt.figure(figsize=(8, 5))
sns.countplot(data=startups_data, x='status', order=startups_data['status'].value_counts().index)
plt.title('Distribuci√≥n de Status de las Startups', fontsize=16)
plt.xlabel('Status', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.tight_layout()
plt.show()

# Principales categor√≠as por cantidad
category_counts = startups_data['category_list'].value_counts().head(10)
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Categor√≠as de Startups', fontsize=16)
plt.xlabel('Categor√≠as', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Distribuci√≥n geogr√°fica: principales pa√≠ses
country_counts = startups_data['country_code'].value_counts().head(10)
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='green')
plt.title('Top 10 Pa√≠ses con m√°s Startups', fontsize=16)
plt.xlabel('Pa√≠s', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.tight_layout()
plt.show()

# Relaci√≥n entre funding_total_usd y status
plt.figure(figsize=(10, 6))
sns.boxplot(data=startups_data, x='status', y='funding_total_usd')
plt.yscale('log')  # Escala logar√≠tmica para mejor visualizaci√≥n
plt.title('Distribuci√≥n de Financiaci√≥n Total por Status', fontsize=16)
plt.xlabel('Status', fontsize=12)
plt.ylabel('Financiaci√≥n Total (USD)', fontsize=12)
plt.tight_layout()
plt.show()

# Correlaci√≥n entre variables num√©ricas
numerical_columns = startups_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = startups_data[numerical_columns].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlaci√≥n entre Variables Num√©ricas', fontsize=16)
plt.tight_layout()
plt.show()

# Crear un gr√°fico de barras agrupado para 'status' y 'category_list'
main_categories = category_counts.index
filtered_data = startups_data[startups_data['category_list'].isin(main_categories)]
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_data, y='category_list', hue='status', order=main_categories)
plt.title('Distribuci√≥n de Status por Principales Categor√≠as', fontsize=16)
plt.xlabel('Cantidad', fontsize=12)
plt.ylabel('Categor√≠a', fontsize=12)
plt.legend(title='Status', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {
    'cat_var1': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'] * 10,
    'cat_var2': ['X', 'Y', 'Z', 'X', 'Y', 'Z', 'X', 'Y', 'Z', 'X'] * 10,
    'cont_var': range(100)
}
df = pd.DataFrame(data)

# Seleccionar los primeros 20 valores √∫nicos de 'cat_var1'
top_categories = df['cat_var1'].unique()[:20]
filtered_df = df[df['cat_var1'].isin(top_categories)]

# Crear el gr√°fico de barras agrupado por las categor√≠as
plt.figure(figsize=(12, 6))
sns.barplot(data=filtered_df, x='cat_var1', y='cont_var', hue='cat_var2', ci=None)
plt.title("Gr√°fico de Barras para las Primeras 20 Categor√≠as de 'cat_var1'")
plt.xlabel("Categor√≠a (cat_var1)")
plt.ylabel("Valor Continuo Promedio (cont_var)")
plt.legend(title="Categor√≠a (cat_var2)")
plt.xticks(rotation=45)
plt.show()
```
